```python
from pyspark import SparkContext, SparkConf

conf = SparkConf().setAppName("Lab1").setMaster("local[*]")
sc = SparkContext(conf=conf)

raw_rdd = sc.textFile("Lab1/data/sample.txt") # Suppose root where you called `uv run jupyter lab` was DistrArch/
```

```python
raw_rdd.take(10)
```

```python
filtered_rdd = raw_rdd \
    .filter(lambda line: line.startswith("ho"))
```

```python
maxfreq = filtered_rdd.top(1, lambda line: int(line.split('\t')[1]))
```

```python
threshold = 0.8 * int(maxfreq[0].split('\t')[1])
top_words_rdd = filtered_rdd.filter(lambda line : int(line.split('\t')[1]) > threshold)
```

```python
# Debug
print(f"{maxfreq=}")
print(f"{top_words_rdd.count()=}")
```

```python
top_words_rdd.map(lambda p: p.split('\t')[0]).saveAsTextFile("Lab1/output/result")
```
